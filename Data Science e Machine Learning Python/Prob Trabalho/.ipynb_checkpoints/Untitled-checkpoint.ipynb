```python
count = df["Gols"].count()
count # Soma total de todos os gols
frequencia = df["Gols"].value_counts()
frequencia_relativa = frequencia / count
frequencia_acumulada = frequencia_relativa.cumsum()
percentual = df['Gols'].value_counts(normalize = True)*100
dist_freq_quantitativas = pd.DataFrame({'Frequência Absoluta': frequencia, 'Frequência Relativa': frequencia_relativa, 'Frequência Acumulada': frequencia_acumulada, 'Porcentagem(%)': percentual}, index=df["Gols"].unique())
dist_freq_quantitativas.sort_index(ascending=True)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
  
tecnicos = ['Cuca (Atlético Mg)', 'Renato Gaucho (Flamengo)', 'Abel Ferreira (Palmeiras)',  
        'Voj Voda (Fortaleza)', 'Silvinho (Corinthians)'] 

sum_Atletico = sum(Atlético_Mg)
sum_Flamengo = sum(Flamengo)
sum_palmeiras = sum(Palmeiras)
sum_Fortaleza = sum(Fortaleza)
sum_corintians = sum(Corinthians)

soma_total = sum_Atletico + sum_Flamengo + sum_palmeiras + sum_Fortaleza + sum_corintians
soma_total

porcent_Atletico = (sum_Atletico / soma_total) * 100 
porcent_Flamengo = (sum_Flamengo / soma_total) * 100
porcent_Palmeiras = (sum_palmeiras / soma_total) * 100
porcent_Fortaleza = (sum_Fortaleza / soma_total) * 100
porcent_Corinthians = (sum_corintians / soma_total) * 100
  
    
data = [porcent_Atletico, porcent_Flamengo, porcent_Palmeiras, porcent_Fortaleza, porcent_Corinthians]
  
  
explode = (0.2, 0.1, 0.1, 0.1, 0.0) 
colors = ( "orange", "cyan", "brown", 
          "grey", "indigo") 
wp = { 'linewidth' : 1, 'edgecolor' : "green" } 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} g)".format(pct, absolute) 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data), 
                                  explode = explode,  
                                  labels = tecnicos, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="magenta")) 
ax.legend(wedges, tecnicos, 
          title ="Técnicos", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Customizing pie chart") 
# plt.show() 
```
